````python
# you may use the functions below

def arange(i: int):
  "This can often replace for-loops."
  return torch.tensor(range(i))

def where(q, a, b):
  """
  This can often replace if-statements.
  It can be used (for each element of the tensors) as:
    if q:
      a
    else:
      b
  """
  return (q * a) + (~q) * b

def ones(i: int):
  "This returns a vector of all ones."
  return (arange(i) == arange(i))


# example usages

example_arange = arange(6)
print("arange example:", example_arange, "\n")

q_example = torch.tensor([False, True])
a_example = torch.tensor([1])
b_example = torch.tensor([-10, 0])
print("q_example: ", q_example)
print("a_example: ", a_example)
print("b_example: ", b_example)
return_example = where(q_example, a_example, b_example)
print("return:    ", return_example, "\n")


print("ones:", ones(6))
````
